# Parse GENCODE Mouse GTF to BED12 + Metadata

1. Loads GENCODE mouse GTF using pyrion
2. Saves complete BED12 + metadata for all transcripts
3. Saves quick test subset (chrX, chr7, chr19)

```python
from pathlib import Path
import sys

pyrion_path = Path("/Users/Bogdan.Kirilenko/PycharmProjects/pyrion")
if pyrion_path not in [Path(p) for p in sys.path]:
    sys.path.insert(0, str(pyrion_path))

from pyrion.io import read_gtf
```

## 1. Load GENCODE GTF

```python
gtf_path = Path("gencode.vM38.annotation.gtf")
transcripts = read_gtf(gtf_path)
print(f"Loaded {len(transcripts):,} transcripts from {gtf_path}")
```

## 2. Save complete annotation

```python
bed_path = Path("mm39_gencode_all_transcripts.bed")
meta_path = Path("mm39_gencode_metadata.tsv")

transcripts.sort(by="position").save_to_bed12(str(bed_path))
transcripts.save_biodata(meta_path)

print(f"Saved {len(transcripts):,} transcripts")
print(f"  BED12:    {bed_path} ({bed_path.stat().st_size / 1e6:.1f} MB)")
print(f"  Metadata: {meta_path} ({meta_path.stat().st_size / 1e6:.1f} MB)")
```

```python
## 3. Quick test subset (chrX, chr7, chr19)
```

```python
QUICK_TEST_CHROMS = {"chrX", "chr7", "chr19"}

quick_test = transcripts.filter_by_chroms(QUICK_TEST_CHROMS).sort(by="position")

qt_bed_path = Path("mm39_quick_test_transcripts.bed")
qt_meta_path = Path("mm39_quick_test_metadata.tsv")

quick_test.save_to_bed12(str(qt_bed_path))
quick_test.save_biodata(qt_meta_path)

print(f"Quick test: {len(quick_test):,} transcripts on {', '.join(sorted(QUICK_TEST_CHROMS))}")
print(f"  BED12:    {qt_bed_path} ({qt_bed_path.stat().st_size / 1e6:.1f} MB)")
print(f"  Metadata: {qt_meta_path} ({qt_meta_path.stat().st_size / 1e6:.1f} MB)")
```

## Output files

- `mm39_gencode_all_transcripts.bed` — full BED12 annotation
- `mm39_gencode_metadata.tsv` — transcript_id, gene_id, biotype, gene_name
- `mm39_quick_test_transcripts.bed` — BED12 for chrX, chr7, chr19 only
- `mm39_quick_test_metadata.tsv` — metadata for chrX, chr7, chr19 only
